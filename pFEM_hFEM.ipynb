```python
import numpy as np
import matplotlib.pyplot as plt
def gauss_quadrature(f, a, b, n=2):
    """
    计算高斯积分
    """
    x, w = np.polynomial.legendre.leggauss(n)
    gauss = 0.5*(b - a)*np.sum(w * f(0.5*(b - a)*x + 0.5*(b + a)))
    return gauss

def local_stiffness_matrix(a, b):
    """
    计算局部刚度矩阵
    """
    h = b - a
    local_matrix = np.array([[1, -1], [-1, 1]]) / h
    return local_matrix

def local_load_vector(f, a, b):
    """
    计算局部载荷向量
    """
    h = b - a
    integral_0 = gauss_quadrature(lambda x: f(x) * (1 - (x - a) / h), a, b)
    integral_1 = gauss_quadrature(lambda x: f(x) * (x - a) / h, a, b)
    return np.array([integral_0, integral_1])

def assemble_global(A_local, F_local, A_global, F_global, element_dofs):
    """
    组装全局刚度矩阵和全局载荷向量
    """
    for i in range(2):
        F_global[element_dofs[i]] += F_local[i]
        for j in range(2):
            A_global[element_dofs[i], element_dofs[j]] += A_local[i, j]
    return A_global, F_global

def finite_element_1d_poisson(f, a, b, alpha, beta, num_elements):
    """
    一维有限元方法解决泊松方程
    """
    # 计算节点位置和元素
    h = (b - a) / num_elements
    nodes = np.linspace(a, b, num_elements + 1)
    elements = [[i, i + 1] for i in range(num_elements)]

    # 初始化全局刚度矩阵和全局载荷向量
    A_global = np.zeros((num_elements + 1, num_elements + 1))
    F_global = np.zeros(num_elements + 1)

    # 循环每个元素，计算并组装全局刚度矩阵和全局载荷向量
    for element in elements:
        element_dofs = element
        a_local = nodes[element[0]]
        b_local = nodes[element[1]]
        A_local = local_stiffness_matrix(a_local, b_local)
        F_local = local_load_vector(f, a_local, b_local)
        A_global, F_global = assemble_global(A_local, F_local, A_global, F_global, element_dofs)

    # 应用边界条件
    F_global[0] -= alpha * A_global[0, 0]
    F_global[-1] -= beta * A_global[-1, -1]
    A_global[0, :] = 0
    A_global[-1, :] = 0
    A_global[0, 0] = 1
    A_global[-1, -1] = 1
    F_global[0] = alpha
    F_global[-1] = beta

    # 求解线性系统
    U = np.linalg.solve(A_global, F_global)
    return nodes, U

# 测试函数
def f(x):
    return np.sin(np.pi*x)

# 定义边界和边界条件
a, b = 0, 1
alpha, beta = 0, 0
num_elements = 5

# 解决泊松方程
nodes, U = finite_element_1d_poisson(f, a, b, alpha, beta, num_elements)

# 打印结果
print(U)
plt.plot(U)
```
